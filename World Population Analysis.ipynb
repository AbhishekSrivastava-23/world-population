Adding column to display **population in millions**, for better understability.
<br>
Very old population data is not of much use, hence **removing the column for 1970 Population**. ðŸ‘‡

```python
df["Population(in M)"] = df["2022 Population"]/1000000 #Adding
df.drop(['1970 Population'], axis = 1, inplace = True) #Removing
df
```

---
CONTINENT WISE ANALYSIS

```python
continents = df['Continent'].value_counts()
plt.pie(continents, labels = continents.index, explode = (0.1, 0.1, 0.1, 0.1, 0.2, 0.2), wedgeprops = {"edgecolor": "black"})
plt.title("NUMBER OF COUNTRIES IN DIFFERENT CONTINENTS")
plt.show()
continents
```

ðŸ‘‰ Less countries are located in **Oceania** and **South America**.

```python
# Population of different continents
cdf = df.groupby("Continent")
pop_sum = cdf["Population(in M)"].sum()
plt.pie(pop_sum, labels = pop_sum.index, explode = (0.1, 0.1, 0.1, 0.1, 0.5, 0.1), wedgeprops = {"edgecolor": "black"})
plt.title("POPULATION OF DIFFERENT CONTINENTS (In Millions)")
plt.show()
pop_sum
```

ðŸ‘‰ Though **Asia** is comparable with other continents in number of countries, it **has a Lion's share in terms of population**.
<br>
ðŸ‘‰ **Oceania has negligible population**, despite having a noticeable presence on the chart for number of countries.

```python
# Percentage of Population of Asian Countries in the World
df[df['Continent']=='Asia']['Population(in M)'].sum()/df['Population(in M)'].sum()*100
```

ðŸ‘‰ *Asian* countries account for a whooping â‰ˆ**59.2%** of the World's population.

---

POPULATION OF INDIA YEAR-WISE ðŸ‘‡

```python
sel_ind = df[df["Country"]=="India"]
ind_pop = sel_ind.iloc[:,11:4:-1].values.tolist()[0]
for i in range(7):
  ind_pop[i] /= 1e7
plt.plot([1980, 1990, 2000, 2010, 2015, 2020, 2022], ind_pop)
plt.xlabel("Year")
plt.ylabel("Population (in crores)")
plt.title("POPULATION OF INDIA YEAR-WISE")
plt.show()
```

ðŸ‘‰ The population of India has grown over the years at a steady rate â†—

---

POPULATION OF TOP 5 COUNTRIES YEAR-WISE ðŸ‘‡

```python
# Top 5 most populous countries
top5 = df.sort_values(by="Population(in M)", ascending=False) #Sorting by population
indx = top5['Country'].head(5)
pop2022 = top5['2022 Population'].head(5)
pop2010 = top5['2010 Population'].head(5)
pop1980 = top5['1980 Population'].head(5)
plt.bar(indx, pop2022/1e6, label = "2022 Population", edgecolor = "black")
plt.bar(indx, pop2010/1e6, label = "2010 Population", edgecolor = "black")
plt.bar(indx, pop1980/1e6, label = "1980 Population", edgecolor = "black")
plt.ylabel("Population (in millions)")
plt.title("POPULATION OF TOP 5 COUNTRIES YEAR-WISE")
plt.legend()
plt.show()
top5.head(5)
```

ðŸ‘‰ **China** and **India** are way ahead of other countries in terms of population.<br>
ðŸ‘‰ *India* gained a huge population **between 1980 and 2010**.
